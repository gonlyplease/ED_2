```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from tqdm import tqdm
from scipy.sparse import load_npz
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MaxAbsScaler
from sklearn.metrics import fbeta_score, accuracy_score, precision_score, recall_score
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

## Loading the data

```python
y_train = pd.read_csv('./data/processed/train_set.csv').set_index('Unnamed: 0')['sensitive']
y_valid = pd.read_csv('./data/processed/valid_set.csv').set_index('Unnamed: 0')['sensitive']
y_test  = pd.read_csv('./data/processed/test_set.csv').set_index('Unnamed: 0')['sensitive']

X_train = load_npz('./data/processed/tfidf_features_train.npz')
X_valid = load_npz('./data/processed/tfidf_features_valid.npz')
X_test  = load_npz('./data/processed/tfidf_features_test.npz')
```

```python
scaler = MaxAbsScaler()

X_tr_sc = scaler.fit_transform(X_train)
X_vd_sc = scaler.transform(X_valid)
X_ts_sc = scaler.transform(X_test)

X_tr_sc
```

## Hyperparameter grid search

```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'max_iter': [10, 50, 100],
    'penalty': ['l2', None],
    'class_weight': [None, "balanced"],
    'solver': ['lbfgs', ]
}

# Storing results
res= []

for C in tqdm(param_grid['C']):
    for max_iter in tqdm(param_grid['max_iter']):
        for penalty in param_grid['penalty']:
          for weights in param_grid['class_weight']:
            for solver in param_grid['solver']:
              estimator = LogisticRegression(max_iter=max_iter, C=C, penalty=penalty, class_weight=weights, solver=solver, random_state=1111)
              estimator.fit(X_tr_sc, y_train)
              res.append([C, max_iter, penalty, weights, fbeta_score(y_valid, estimator.predict(X_vd_sc), beta=1)])
```

```python
res = pd.DataFrame(res)
res.columns=['C', 'max_iter', 'penalty', 'class_weight', 'f1']
print(res.loc[res['f1'].idxmax(), :])
```

```python
res
```

Best model:
* C: seems not to matter much, let's leave it at default of 1
* max_iter: 100 is enough
* penalty: None is better
* class_weight: None is better

## Testing the model on test data and calculating all the metrics

```python
estimator = LogisticRegression(max_iter=50, C=10, penalty="l2", random_state=1111)
estimator.fit(X_tr_sc, y_train)
pred = estimator.predict(X_ts_sc)

results = pd.DataFrame(columns=['Precision', 'Recall', 'F-1', 'F-2', 'Accuracy'])
results.loc[0, 'Precision'] = precision_score(y_test, pred)
results.loc[0, 'Recall'] = recall_score(y_test, pred)
results.loc[0, 'F-1'] = fbeta_score(y_test, pred, beta=1)
results.loc[0, 'F-2'] = fbeta_score(y_test, pred, beta=2)
results.loc[0, 'Accuracy'] = accuracy_score(y_test, pred)

print('Resulting model performance:')
results
```
